```python
df_countries = df[(df["location"].isin(["Italy", "Germany", "France"])) & 
                  (df["date"] >= "2022-05-01") & (df["date"] <= "2023-04-30")]
```

```python
df_countries_cleaned = df_countries.dropna(subset=["icu_patients"])
```

```python
df_countries_cleaned
```

```python
import seaborn as sns
```

```python

sns.boxplot(data=df_countries_cleaned, x="location", y="icu_patients", palette="Set2")
```

In Francia, il numero di pazienti in terapia intensiva si concentra intorno ai 1000, con la maggior parte dei valori tra 750 e 1250. Ci sono alcuni valori estremi, ma senza outlier significativi.

La Germania ha una mediana simile, ma con maggiore variabilità: i pazienti oscillano tra 500 e 1300-1400, con picchi fino a 1800. Questo indica una distribuzione più ampia rispetto alla Francia.

L’Italia, invece, ha numeri nettamente inferiori. La mediana è sotto i 300, con valori tra 150 e 350 e una distribuzione più compatta.

In sintesi, la Germania mostra la maggiore variabilità, la Francia ha una distribuzione più stabile, mentre l’Italia registra il numero più basso di pazienti e meno fluttuazioni.
